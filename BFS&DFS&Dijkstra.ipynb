# BFS & DFS & Dijkstra

## Queue

```python
class Queue:
    max_size = 10
    def __init__(self, max_size):
        self.items = []
        self.max_size = max_size
        size = 0

    def isFull(self):
        return len(self.items) == self.max_size
    def isEmpty(self):
        return len(self.items) == 0
    def enqueue(self, e):
        if self.isFull():
            return "Error, Queue is full!"
        else:
            self.items.append(e)
    def dequeue(self):
        if self.isEmpty():
            return "Error, Queue is empty!"
        else:
            return self.items.pop(0)
        return e
    def set_maxSize(self, s):
        self.max_size = s
    def front(self):
        if self.isEmpty():
            return "Error, Queue is empty!"
        else:
            print(self.items[0])
    def rear(self):
        if self.isEmpty():
            return "Error, Queue is empty!"
        else:
            print(self.items[-1])

    def size(self):
        return len(self.items)
    
    def __str__(self):
        return "Queue = {}".format(self.items)
```

## Stack

```python
class Stack:
    max_size = 10
    def __init__(self, max_size):
        self.items = []
        self.max_size = max_size
        size = 0

    def isFull(self):
        return len(self.items) == self.max_size
    
    def isEmpty(self):
        return len(self.items) == 0
    
    def push(self, e):
        if self.isFull():
            return "Error, Stack is full!"
        else:
            self.items.append(e)
    def pop(self):
        if self.isEmpty():
            return "Error, Stack is empty!"
        else:
            return self.items.pop()
        return e
    def set_maxSize(self, s):
        self.max_size = s
    def front(self):
        if self.isEmpty():
            return "Error, Stack is empty!"
        else:
            print(self.items[0])
    def rear(self):
        if self.isEmpty():
            return "Error, Stack is empty!"
        else:
            print(self.items[-1])

    def size(self):
        return len(self.items)
    
    def __str__(self):
        return "Stack = {}".format(self.items)
```

## BFS

```python
def bfs(graph, root):
    q = Queue(len(graph.keys()))
    visited = set()
    
    visited.add(root)
    q.enqueue(root)
    
    while q.isEmpty() == False:
        n = q.dequeue()
        print (n, end = " ") 
        
        for x in graph[n]:
            if x not in visited:
                visited.add(x)
                q.enqueue(x)
    
```

## DFS

```python
def dfs(graph, root):
    s = Stack(len(graph.keys()))
    visited = set()
    
    visited.add(root)
    s.push(root)
    
    
    while s.isEmpty() == False:
        n = s.pop()
        print (n, end = " ") 

        for x in graph[n][::-1]:
            if x not in visited:
                visited.add(x)
                s.push(x)
    
```

## Dijkstra

```python
def dijkstra(graph, root):
  inf = 10000
  nodes = sorted(graph.keys())
  unvisited = {node: inf for node in nodes} #using None as +inf
  visited = {}
  current = root
  currentDistance = 0
  unvisited[current] = currentDistance

  while True:
      for neighbour, distance in graph[current].items():
          if neighbour not in unvisited: continue
          newDistance = currentDistance + distance
          if unvisited[neighbour] is inf or unvisited[neighbour] > newDistance:
              unvisited[neighbour] = newDistance
      visited[current] = currentDistance
      del unvisited[current]
      if not unvisited: break
      candidates = [node for node in unvisited.items() if node[1]]
      current, currentDistance = sorted(candidates, key = lambda x: x[1])[0]

  print(visited)
```
